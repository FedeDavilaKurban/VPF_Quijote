```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
import os

dir = '/home/federico/Proyectos/VPF_Quijote/data/output/2pcf/fiducial/'

files = os.listdir(dir)

xi0 = []
xi2 = []
xi4 = []
for file in files:
    data = ascii.read(dir+file)
    y0 = data['xi0']
    y2 = data['xi2']
    y4 = data['xi4']
    x = data['s']
    plt.plot(x,y0*x**2,c='C00',alpha=.01)
    plt.plot(x,-y2*x**2,c='C01',alpha=.01)
    plt.plot(x,y4*x**2,c='C02',alpha=.01)

    xi0.append(y0)
    xi2.append(y2)
    xi4.append(y4)

y0mean = np.mean(xi0,axis=0)
y2mean = np.mean(xi2,axis=0)
y4mean = np.mean(xi4,axis=0)
plt.plot(x,y0mean*x**2,c='k')
plt.plot(x,-y2mean*x**2,c='k')
plt.plot(x,y4mean*x**2,c='k')
plt.ylim([-75,75])
plt.show()
```
